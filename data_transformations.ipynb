### LET'S CREATE A GROUP ID WITH GROUP DEFINED BY UNIQUE CYL AND GEAR VALUES

```r
# check the # of unique combinations of cyl and gear
mtcars%>%
arrange(cyl,gear)%>%
distinct(cyl,gear)

# We have 8 groups. curr_group_id assigned a group id based on the grouping vars
mtcars%>%
group_by(cyl,gear)%>%
mutate(group_id = cur_group_id())%>%
arrange(group_id)%>%
distinct(group_id)

mtcars%>%
group_by(cyl,gear)%>%
mutate(group_id = cur_group_id())%>%
arrange(group_id)%>%
head()
```
